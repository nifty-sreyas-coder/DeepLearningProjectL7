# Data Pre Processing + Encoding

```python
import numpy as np
import pandas as pd 
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv("adult.csv")
df 
```

```python
df.columns=['age', 'workclass', 'fnlwgt', 'education', 'education-num',
       'marital-status', 'occupation', 'relationship', 'race', 'sex',
       'capital-gain', 'capital-loss', 'hours-per-week', 'native-country','income']

df
```

```python
pd.Series(df.workclass)[0]
```

```python
df[df.workclass == " ?"]
```

```python
Allcols = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
       'marital-status', 'occupation', 'relationship', 'race', 'sex',
       'capital-gain', 'capital-loss', 'hours-per-week', 'native-country','income']

df = df.replace(" ?", np.nan)

df[df.workclass == " ?"]
```

```python
df.isna().sum()
```

```python
encoder = OrdinalEncoder(encoded_missing_value=np.nan)
```

```python
columnsToEncode =['workclass', 'marital-status', 'occupation', 'relationship', 'race', 'sex','native-country','income']

for value in columnsToEncode:
    df[value] = encoder.fit_transform(df[[value]])
    
```

```python
df
```

```python
df = df.drop("education", axis=1)

df
```

```python
df.isna().sum()
```

```python
containsMissingValues = ["workclass", "occupation", "native-country"]

for item in containsMissingValues:
    df[item].fillna(df[item].mode()[0], inplace=True)

df
```

```python
df.isna().sum()
```

```python
df = df.drop("fnlwgt", axis=1)

df
```

# Training and Testing

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

scalledArray = (scaler.fit_transform(df))

Scalled_df = pd.DataFrame(scalledArray,  columns=df.columns)
Scalled_df
```

```python
x = Scalled_df.drop("income", axis=1)
x
```

```python
y = Scalled_df["income"]
y
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=Scalled_df["income"])
```

```python
model = DecisionTreeClassifier(ccp_alpha=0.01, class_weight={0.0: 1, 1.0: 3})
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
model.score(x_test, y_test)
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout, InputLayer, BatchNormalization
from tensorflow.keras.optimizers import Adam
```

```python
y_train
```

```python
Sequential
```

```python
model2 = Sequential([
    Dense(256, activation="relu", input_shape=(12, )),
    BatchNormalization(),
    Dropout(0.35),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.35),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.25),
    Dense(32, activation='relu'),
    BatchNormalization(),
    Dense(1, activation="sigmoid")
])
```

```python
model2.compile(optimizer=Adam(learning_rate=0.0007),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model2.fit(x_train, y_train, epochs=40, validation_split=0.2)
```

```python

tensorflow.keras.models.save_model(model2, "model.h5")
```
